1. Você trabalha em uma transportadora urbana que opera 3 tipos de veículos urbanos de carga(VUC):

  - O modelo 1, da marca JAC, tem 38 m3 de volume útil e consumo de 6 km/l;

  - O modelo 2, da marca KIA, tem 44 m3 de volume útil e consumo de 9 km/l;

  - O modelo 3, da marca Mercedes-Benz, tem 42 m3 de volume útil e consumo de 7 km/

A empresa realiza somente fretes cheios, ou seja, com o VUC com carga próxima a máxima permitida. A transportadora oferece caixas de transporte com as seguintes dimensões:

  - Caixa modelo A com 0,5 m3 de capacidade;

  - Caixa modelo B com 1,0 m3 de capacidade;

  - Caixa modelo C com 2,0 m3.

Elabore um algoritmo em Python que receba (entrada de dados) a quantidade de cada tipo de caixa a ser transportada e a distância a ser percorrida em quilômetros. O algoritmo deve determinar (processamento de dados) o volume total a ser transportado e escolher o caminhão (if...elif...else) que pode fazer a entrega gastando a menor quantidade de combustível e deve
apresentar (saída) o modelo de caminhão escolhido, o volume total a ser transportado e o gasto esperado de combustível.

```python
#Entrada de dados:

#Entrada da distancia:
Distancia = float(input("Digite a distancia que deve ser percorrida em Km: "))

#Entrada de volume:
print("Agora, digite a quantidade de caixas de acordo com o volume, sendo elas de 0,5 m3, 1 m3 e 1,5 m3:\nLevar em conta também que os tipos de caminhões comportam 38, 42, 44 m3 respectivamentes.")

qt_caixa_A = int(input("Caixa A (0,5 m3 de capacidade): "))
qt_caixa_B = int(input("Caixa B (1 m3 de capacidade): "))
qt_caixa_C = int(input("Caixa C (1,5 m3 de capacidade): "))

#Processamento de volume:
volume_A = qt_caixa_A * 0.5
volume_B = qt_caixa_B * 1
volume_C = qt_caixa_C * 1.5

volume_total = volume_A + volume_B + volume_C

#Saída de dados e processamento de distancia
if volume_total <= 38:
  print(f"Tipo de caminhão: modelo 1, marca JAC; volume total: 38 m3, sendo o volume transportado: {volume_total} m3; E sendo a distancia de {Distancia}, gastando assim a quantidade de combustivel de {Distancia / 6} litros.")

elif volume_total <= 42:
  print(f"Tipo de caminhão: modelo 3, marca Mercedes-Benz; volume total: 42 m3, sendo o volume transportado: {volume_total} m3; E sendo a distancia de {Distancia}, gastando assim a quantidade de combustivel de {Distancia / 7} litros.")

else:
  print(f"Tipo de caminhão: modelo 2, marca KIA; volume total: 44 m3, sendo o volume transportado: {volume_total} m3; E sendo a distancia de {Distancia}, gastando assim a quantidade de combustivel de {Distancia / 9} litros.")
```

líquido, baseado somente nos descontos de INSS, IRRF e dependentes. As tabelas de INSS e IRRF são
mostradas abaixo. Considerar o desconto integral do INSS na faixa salarial, ao invés de um desconto
progressivo. Os valores exibidos devem estar no formato moeda brasileiro com 2 casas decimais.

```python
SalarioBruto = float(input("Entre com seu sálario bruto: "))
NumDependentes = int(input("Entre com o número de dependentes: "))

#PROCESSAMENTO DE DADOS

# 1. Cálculo do INSS

if SalarioBruto > 0 and SalarioBruto <= 1320:
  InssFaixa1 = SalarioBruto * 0.075
  InssDescontado = InssFaixa1
elif SalarioBruto >= 1320.01 and SalarioBruto <= 2571.29:
  InssFaixa1 = 1320 * 0.075
  InssFaixa2 = (SalarioBruto - 1320) * 0.09
  InssDescontado = InssFaixa1 + InssFaixa2
elif SalarioBruto >= 2571.30 and SalarioBruto <= 3856.94:
  InssFaixa1 = 1320 * 0.075
  InssFaixa2 = (2571.29 - 1320) * 0.09
  InssFaixa3 = (SalarioBruto - 2571.29) * 0.12
  InssDescontado = InssFaixa1 + InssFaixa2 + InssFaixa3
elif SalarioBruto >= 3856.95 and SalarioBruto <= 7507.49:
  InssFaixa1 = 1320 * 0.075
  InssFaixa2 = (2571.29 - 1320) * 0.09
  InssFaixa3 = (3856.95 - 2571.29) * 0.12
  InssFaixa4 = (SalarioBruto - 3856.94) * 0.14
  InssDescontado = InssFaixa1 + InssFaixa2 + InssFaixa3 + InssFaixa4
elif SalarioBruto >= 7507.50:
  InssDescontado = 1051.04


# 2. Cálculo do salario Base
SalarioBase = SalarioBruto - InssDescontado - (NumDependentes * 189.59)

# 3. Cálculo do IRRF
if SalarioBase >= 0 and SalarioBase <= 2112:
  IrrfDescontado = 0
elif SalarioBase >= 2112 and SalarioBase <= 2826.65:
  IrrfDescontado = (SalarioBase * 0.075) - 158.40
elif SalarioBase >= 2826.66 and SalarioBase <= 3751.05:
  IrrfDescontado = (SalarioBase * 0.15) -370.40
elif SalarioBase >= 3751.06 and SalarioBase <= 4664.68:
  IrrfDescontado = (SalarioBase * 0.225) - 651.73
elif SalarioBase >= 4664.68:
  IrrfDescontado = (SalarioBase * 0.275) - 884.96

# 4. Cálculo do Salário Líquido

SalarioLiquido = SalarioBruto - InssDescontado - IrrfDescontado

# SAÍDA DE DADOS

print("  ")
print(f"O desconto do INSS é  R$ {InssDescontado:.2f}")
print(f"O valor do salário base é  R$ {SalarioBase:.2f}")
print(f"O valor do IRRF é  R$ {IrrfDescontado:.2f}")
print(f"O valor do salário liquido é  R$ {SalarioLiquido:.2f}")
```
